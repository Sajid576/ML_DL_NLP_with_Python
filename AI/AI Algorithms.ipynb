that a numeric key is followed by a tuple of diverse elements separated by ‘\t’ in a line. The second method should return the
list of the 1
st elements of the tuples in the dictionary. You need also to arrange use of the module with an entry for the name of
the file.

```
dictionary={}

def readFile():
    file = open("/content/drive/MyDrive/My Colab Notebook/final2.txt", "r")

    Lines = file.readlines()


    for line in Lines:
        x=line.split('\t')
        key=x[0]
        lst=list(x)
        lst.pop(0)
        myTuple=tuple(lst)
        dictionary[key] = myTuple 

    print(dictionary)

def getFrstElem():
    for key,val in dictionary.items():
        print(key+'-->'+str(val[0]))

readFile()
getFrstElem()
```

Write a Python module that has two methods, one for creating a file with n lines, and another for displaying the average of the
numeric parts of the first m lines of the file created. Assume that each line of the file contains one sting entry followed by one
number entry separated by ‘;’. You need also to arrange use of the module with entries for the name of the file and the values
of m and n.


```
import random


def craeteFile(n):
    f = open("/content/drive/MyDrive/My Colab Notebook/final1.txt", "w")
    while(n!=0):
        lineLen=random.randrange(1, 10, 3)
        for i in range(lineLen):
            num=random.randrange(1, 10, 1)
            f.write(str(num))
            if(i!=lineLen-1):
                f.write(';')
        f.write('\n')   
        n-=1
    f.close()
    
def FindAvg(m):
      f = open("/content/drive/MyDrive/My Colab Notebook/final1.txt", "r")
      Lines = f.readlines()
      den=0
      numerator=0
      for line in Lines:
         if(m==0):
            break
         else:
            x=line.split(';')
            lst=list(map(int,x))
            numerator+=sum(lst)
            den+=len(lst)
            m-=1
      AVG=numerator/den
      return AVG





n = int(input('n: '))
craeteFile(n)
m = int(input('m: '))


print('Average of m lines: '+str(FindAvg(m)))
```

```
# Greedy Best First Search

from queue import PriorityQueue

graph = {

    'a': [['b', 46], ['c', 54]],

    'b': [['c', 47], ['d', 36],['e',25]],

    'd': [['e',68], ['g', 62]],

    'c': [['e',35],['f',32]],

    'e': [['f',57],['g',58]],

    'f':[['g',45]]
    
    }

heuristic_values = { 'a': 75, 'b': 51, 'c': 53, 'd': 56, 'e': 54,'f':38, 'g': 0}

pq = PriorityQueue()

pp = []
visited = []
source = 'a'
destination = 'g'

pq.put( ( heuristic_values[source] , source,0) )

while True:
    print("p.pqueue: ")
    print(pq.queue)
    print("possible path: ")
    print(pp)
    pathAvailable=False
    item = pq.get()    
    pp.append((item[1],item[2] ))   # node,cost
    currentNode=item[1]
    visited.append(currentNode)
    #if destination node is found
    if(currentNode==destination):
        break;

    for neighbour in graph[currentNode]:
       
        if(neighbour[0] not in visited):
            # heuristic value of neighbour,node,cost from current to neighbour --> pq
            pq.put(( heuristic_values[ neighbour[0] ] , neighbour[0], neighbour[1]) ) 
            pathAvailable=True
    
    #if further approach is not possible from a particular node ,then pop last item in the pp
    if(pathAvailable==False):
        pp.pop(-1)        



print("Best possible path using greedy best first search: ")
print(pp)
print('Total cost: ')
total_cost=0;
for item in pp:
    total_cost += item[1]

print(total_cost)
```
